 #### We Need To Remove all the unnecessary parts in the textual data to proceed forward

- URL
- Hashtags
- Mentions
- Special Characters
- Punctuations


```python
import re
def cleanText(txt):
    cleanText = re.sub(r'http\S+\s' , ' ' , txt)
    cleanText = re.sub(r'RT|cc', ' ', cleanText)
    cleanText = re.sub(r'#\S+\s', ' ', cleanText)
    cleanText = re.sub(r'@\S+', '  ', cleanText)  
    cleanText = re.sub(r'[%s]' % re.escape(r"""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub(r'\s+', ' ', cleanText)
    return cleanText
```

```python
cleanText('.\r\n\r\nTools & Technologies: Python, scikit-learn, tfidf, word2vec, doc2vec, cosine similarity, ')
```

```python
# Applying the function to every  Resume in the dataset
dataset['Resume'] = dataset['Resume'].apply(lambda x : cleanText(x))
```

```python
dataset['Resume'] = dataset['Resume'].str.lower()
```

### Vectorization

```python

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(dataset['Resume'])
requiredtext  = tfidf.transform(dataset['Resume'])
```

```python
dataset.head(60)
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(requiredtext,dataset['Category'], test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

## Train the model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
ypred = clf.predict(X_test)
print(accuracy_score(y_test,ypred)*100)
```
